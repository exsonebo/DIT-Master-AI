# Packages importation

```python
# Importation of the data for our classification example
from sklearn.datasets import load_breast_cancer

# Importation of the function to standardize the data
from sklearn.preprocessing import StandardScaler

# Importation of the logistic regression algorithm
from sklearn.linear_model import LogisticRegression

# Importation of the train_test_split function which split randomly our data 
# into a train and test set
from sklearn.model_selection import train_test_split

# Importation of numpy to use of vectors, matrices, tensors.
import numpy as np 

# Importation of the performance metrics
from sklearn.metrics import accuracy_score, precision_recall_curve, f1_score, roc_auc_score, roc_curve, confusion_matrix

# Importation of the package for plot graphics
import matplotlib.pyplot as plt
```

# Data Importation

```python
# Data frame for ou regression example
breast_cancer = load_breast_cancer()
X_classif = breast_cancer.data[:, ]
y_classif = breast_cancer.target
```

Use the Sklearn function *train_test_split* to split your dataset into two random set.

Use a random_state of 123 and use 10% of your dataset for the test set.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html).

```python
# Use the function train_test_split to create your train and test set
#X_train_classif, X_test_classif, y_train_classif, y_test_classif = None

X_train_classif, X_test_classif, y_train_classif, y_test_classif = train_test_split(X_classif, y_classif, test_size=0.2, random_state=42)
```

# Step 2 : Model initialization

In the case of logistic regression, there is no choice of hyperparameter.

It is therefore sufficient to just initialize the function.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

```python
reg = None
```

# Step 3 : Model training 

Use the model named *reg* that you just initialized and train it with the standardized data. 

Hint : use the fonction *fit*

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

```python
# Initialize the logistic regression model
reg = LogisticRegression()

# Train the model with standardized data
reg.fit(X_train_classif_norm, y_train_classif)
```

# Step 4 : Model validation

Your model is now trained, use it to predict the price of the appartment for your training and testing set.

feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html).

```python
y_train_prediction = None

y_test_prediction = None


# Predict class labels for the training set
y_pred_train_classif = reg.predict(X_train_classif_norm)

# Predict class labels for the testing set
y_pred_test_classif = reg.predict(X_test_classif_norm)
```

Use your model to predict the probability for each observation

```python
# Predict probabilities for the training set
y_train_prediction_proba = reg.predict_proba(X_train_classif_norm)

# Predict probabilities for the testing set
y_test_prediction_proba = reg.predict_proba(X_test_classif_norm)
```

Compute the accuracy for the training and the testing set.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html#sklearn.metrics.accuracy_score).

```python
# Compute accuracy score for the training set
acc_train = accuracy_score(y_train_classif, y_pred_train_classif)

# Compute accuracy score for the testing set
acc_test = accuracy_score(y_test_classif, y_pred_test_classif)

print('Accuracy score for the training set:', acc_train)
print('Accuracy score for the testing set:', acc_test)
 
```

Compute the F1 score for the training and the testing set.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html).

```python
# Compute F1 score for the training set
f1_train = f1_score(y_train_classif, y_pred_train_classif)

# Compute F1 score for the testing set
f1_test = f1_score(y_test_classif, y_pred_test_classif)

print('F1 score for the training set:', f1_train)
print('F1 score for the testing set:', f1_test)
```

Compute the AUC for the training and the testing set.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_auc_score.html).

```python
#The AUC score is a metric that ranges from 0 to 1, with higher values indicating better performance. AUC represents the probability that a random positive example is ranked higher than a random negative example. In our case, since we have only two classes, the positive class corresponds to cancer and the negative class corresponds to no cancer.

#The AUC score for the training set is 0.995, which indicates that the model is able to distinguish between positive and negative examples in the training set with high accuracy. The AUC score for the testing set is 0.996, which indicates that the model is able to generalize well to new, unseen data.

# Compute AUC for the training set
auc_train = roc_auc_score(y_train_classif, y_train_prediction_proba[:, 1])

# Compute AUC for the testing set
auc_test = roc_auc_score(y_test_classif, y_test_prediction_proba[:, 1])

print('AUC for the training set:', auc_train)
print('AUC for the testing set:', auc_test) 
```

Compute the confusion matrix.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html). 

```python
# Compute confusion matrix for training set
table_train = confusion_matrix(y_train_classif, y_pred_train_classif)

# Compute confusion matrix for testing set
table_test = confusion_matrix(y_test_classif, y_pred_test_classif)

# Print the results
print('Confusion matrix for the training set : ')
print(table_train)

print('Confusion matrix for the testing set : ')
print(table_test)
```

#Step 5 : Visualization

Use the function roc_curve to compute the true positive rate and false positive rate.

Feel free to use the [doc](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html). 

```python
# Compute true positive rate and false positive rate for training set
fpr_train, tpr_train, _ = roc_curve(y_train_classif, y_train_prediction_proba[:, 1])

# Compute true positive rate and false positive rate for testing set
fpr_test, tpr_test, _ = roc_curve(y_test_classif, y_test_prediction_proba[:, 1])
```

Try to reproduce this figure using matplotlib.

Feel free to use the [doc](https://matplotlib.org/stable/contents.html).

roc_curve (1).png

```python
import matplotlib.pyplot as plt

# plot ROC curve for training set
plt.plot(fpr_train, tpr_train, color='orange', label='Training set')

# plot ROC curve for testing set
plt.plot(fpr_test, tpr_test, color='green', label='Testing set')

# plot diagonal line representing random classifier
plt.plot([0, 1], [0, 1], color='blue', linestyle='--')

# set x and y labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# set title and legend
plt.title('Receiver Operating Characteristic for the breast cancer example')
plt.legend()

# show the plot
plt.show()
```
